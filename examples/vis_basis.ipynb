```python
import numpy as np
import matplotlib.pylab as plt
```

```python
T = np.linspace(0, 1, 4)
func_dict = {
    "func1": lambda x: (x - T[0]) / (T[2] - T[0]) * (x - T[0]) / (T[1] - T[0]),
    "func2": lambda x: (T[2] - x) / (T[2] - T[1]) * (x - T[0]) / (T[2] - T[0]) + (T[3] - x) / (T[3] - T[1]) * (x - T[1]) / (T[2] - T[1]),
    "func3": lambda x: (T[3] - x) / (T[3] - T[1]) * (T[3] - x) / (T[3] - T[2])
}
xs = np.linspace(start=T[0], stop=T[-1], num=100)
res = np.zeros(100)
for idx, x in enumerate(xs):
    if x >= T[0] and x <= T[1]:
        res[idx] = func_dict["func1"](x)
    elif x > T[1] and x <= T[2]:
        res[idx] = func_dict["func2"](x)
    else:
        res[idx] = func_dict["func3"](x)
plt.title("Quadratic B-Spline Blending Function")
plt.plot(xs, res)
plt.xlabel("u")
plt.ylabel("w")
```

```python
T = np.linspace(0, 1, num=5)
func_dict = {
    "func1": lambda x: (x - T[0]) * (x - T[0]) * (x - T[0]) / (T[1] - T[0]) / (T[2] - T[0]) / (T[3] - T[0]),

    "func2": lambda x: (x - T[0]) * (x - T[0]) * (T[2] - x) / (T[2] - T[1]) / (T[3] - T[0]) / (T[2] - T[0]) +
                       (T[3] - x) * (x - T[0]) * (x - T[1]) / (T[2] - T[1]) / (T[3] - T[1]) / (T[3] - T[0]) +
                       (T[4] - x) * (x - T[1]) * (x - T[1]) / (T[2] - T[1]) / (T[3] - T[1]) / (T[3] - T[0]),

    "func3": lambda x: (x - T[0]) * (T[3] - x) * (T[3] - x) / (T[3] - T[2]) / (T[3] - T[1]) / (T[3] - T[0]) +
                       (T[4] - x) * (T[3] - x) * (x - T[1]) / (T[3] - T[2]) / (T[4] - T[1]) / (T[3] - T[1]) +
                       (T[4] - x) * (T[4] - x) * (x - T[2]) / (T[3] - T[2]) / (T[4] - T[2]) / (T[4] - T[1]),

    "func4": lambda x: (T[4] - x) * (T[4] - x) * (T[4] - x) / (T[4] - T[3]) / (T[4] - T[2]) / (T[4] - T[1])
}
xs = np.linspace(start=T[0], stop=T[-1], num=100)
res = np.zeros(100)
for idx, x in enumerate(xs):
    if x >= T[0] and x <= T[1]:
        res[idx] = func_dict["func1"](x)
    elif x > T[1] and x <= T[2]:
        res[idx] = func_dict["func2"](x)
    elif x > T[2] and x <= T[3]:
        res[idx] = func_dict["func3"](x)
    else:
        res[idx] = func_dict["func4"](x)

plt.title("Cubic B-Spline Blending Function")
plt.plot(xs, res)
plt.xlabel("u")
plt.ylabel("w")
```
